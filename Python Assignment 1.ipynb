5.Visualisations using dataset

```python
plt.figure(figsize=(12,6))
plt.scatter(data['PL'],data['PW'])
plt.title('Scatter Diagram')
plt.show
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x= data['SL'], y= data['SW'])
plt.title('Box Plot')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(data['SL'],data['SW'])
plt.title('Bar Graph')
plt.xlabel('SL')
plt.ylabel('SW')
plt.show()
```

```python
plt.hist(data['PL'])
plt.xlabel('PL')
plt.ylabel('Frequency')
plt.show()
```
